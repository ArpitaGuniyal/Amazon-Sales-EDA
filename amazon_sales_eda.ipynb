## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Cleaning and preparing the dataset for analysis

# Create a copy of the original dataset to preserve raw data
df_clean = df.copy()

# Remove the Indian Rupee symbol (₹) and commas from price columns
df_clean['discounted_price'] = (
    df_clean['discounted_price']
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float)
)

df_clean['actual_price'] = (
    df_clean['actual_price']
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float)
)

# Remove the percentage symbol (%) and convert to numeric
df_clean['discount_percentage'] = (
    df_clean['discount_percentage']
    .str.replace('%', '', regex=False)
    .astype(float)
)

# Convert ratings into numeric values
df_clean['rating'] = pd.to_numeric(df_clean['rating'], errors='coerce')

# Remove commas from rating_count and convert to numeric
df_clean['rating_count'] = (
    df_clean['rating_count']
    .str.replace(',', '', regex=False)
)

df_clean['rating_count'] = pd.to_numeric(df_clean['rating_count'], errors='coerce')

# Verify the updated data types
df_clean.info()
````

### What all manipulations have you done and insights you found?

Observation:
* Created a separate copy of the dataset (df_clean) to preserve the original data.
* Converted price columns from string format to numeric by removing the currency symbol (₹) and commas.
* Converted discount_percentage into numeric values by removing the percentage sign.
* Converted rating and rating_count columns to numeric format for statistical analysis.
* The dataset is now suitable for numerical calculations, correlation analysis and visualizations.
* Only the rating_count column contains two missing values, while all other cleaned columns were converted successfully.

#### Chart - 5: Top 10 Highest Rated Products

````python
# Select the top 10 most reviewed products
top_reviewed = (
    df_clean[['product_name', 'rating_count']]
    .dropna()
    .sort_values(by='rating_count', ascending=False)
    .head(10)
)

# Shorten long product names for better readability
top_reviewed['short_name'] = (
    top_reviewed['product_name']
    .str.slice(0, 45) + "..."
)

# Create the figure
plt.figure(figsize=(12, 7))

# Plot horizontal bar chart
plt.barh(
    top_reviewed['short_name'],
    top_reviewed['rating_count'],
    edgecolor='black'
)

# Add chart title and labels
plt.title("Top 10 Most Reviewed Products")
plt.xlabel("Number of Ratings")
plt.ylabel("Product")

# Display the highest value at the top
plt.gca().invert_yaxis()

plt.grid(axis='x', alpha=0.3)

plt.tight_layout()

plt.show()
````

##### 1. Why did you pick the specific chart?

A horizontal bar chart is ideal for comparing products with the highest number of customer ratings. It clearly ranks products by popularity while accommodating long product names.

##### 2. What is/are the insight(s) found from the chart?

* A small number of products have received significantly more customer ratings than others.
* These products are likely among the most popular and widely purchased items on Amazon.
* The gap between the top-reviewed products and the rest suggests that customer attention is concentrated on a limited set of products.

##### 3. Will the gained insights help creating a positive business impact?
