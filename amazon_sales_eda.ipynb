## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Cleaning and preparing the dataset for analysis

# Create a copy of the original dataset to preserve raw data
df_clean = df.copy()

# Remove the Indian Rupee symbol (₹) and commas from price columns
df_clean['discounted_price'] = (
    df_clean['discounted_price']
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float)
)

df_clean['actual_price'] = (
    df_clean['actual_price']
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float)
)

# Remove the percentage symbol (%) and convert to numeric
df_clean['discount_percentage'] = (
    df_clean['discount_percentage']
    .str.replace('%', '', regex=False)
    .astype(float)
)

# Convert ratings into numeric values
df_clean['rating'] = pd.to_numeric(df_clean['rating'], errors='coerce')

# Remove commas from rating_count and convert to numeric
df_clean['rating_count'] = (
    df_clean['rating_count']
    .str.replace(',', '', regex=False)
)

df_clean['rating_count'] = pd.to_numeric(df_clean['rating_count'], errors='coerce')

# Verify the updated data types
df_clean.info()
````

### What all manipulations have you done and insights you found?

Observation:
* Created a separate copy of the dataset (df_clean) to preserve the original data.
* Converted price columns from string format to numeric by removing the currency symbol (₹) and commas.
* Converted discount_percentage into numeric values by removing the percentage sign.
* Converted rating and rating_count columns to numeric format for statistical analysis.
* The dataset is now suitable for numerical calculations, correlation analysis and visualizations.
* Only the rating_count column contains two missing values, while all other cleaned columns were converted successfully.
