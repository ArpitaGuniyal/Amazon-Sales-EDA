## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1: Distribution of Product Ratings

````python
plt.figure(figsize=(8, 5))

plt.hist(
    df_clean['rating'].dropna(),
    bins=15,
    edgecolor='black'
)

plt.title("Distribution of Product Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Products")

plt.grid(alpha=0.3)

plt.show()
````

##### 1.1 Why did you pick the specific chart?

A histogram is ideal for visualizing the distribution of a numerical variable. It helps understand how product ratings are spread across different rating values and whether customers generally give high or low ratings.

##### 1.2 What is/are the insight(s) found from the chart?

* Most products have ratings between 4.0 and 4.5.
* Very few products have ratings below 3.5.
* The dataset is skewed toward highly rated products.

##### 1.3 Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes.
* High ratings indicate that most listed products satisfy customers.
* Businesses can identify the characteristics of highly rated products and replicate those strategies.
* Products with unusually low ratings can be investigated to improve customer satisfaction.

#### Chart - 2: Distribution of Discount Percentage

````python
plt.figure(figsize=(8,5))

plt.hist(
    df_clean['discount_percentage'].dropna(),
    bins=15,
    edgecolor='black'
)

plt.title("Distribution of Discount Percentage")
plt.xlabel("Discount Percentage (%)")
plt.ylabel("Number of Products")

plt.grid(alpha=0.3)

plt.show()
````

##### 2.1 Why did you pick the specific chart?

A histogram is the most appropriate chart for understanding the distribution of a numerical variable. It shows whether products are concentrated around certain discount ranges or spread evenly across different discounts.

##### 2.2 What is/are the insight(s) found from the chart?

* Most products receive discounts between 40% and 60%.
* Very few products have extremely low or extremely high discounts.
* The distribution suggests that Amazon commonly offers moderate discounts rather than very small or exceptionally large discounts.

##### 2.3 Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes.
* Understanding the most common discount range helps sellers design competitive pricing strategies.
* Businesses can benchmark their discounts against the marketplace average.
* Products with unusually high discounts may indicate clearance sales or lower demand and should be monitored to maintain profitability.

#### Chart - 3: Top 10 Product Categories

````python
# Extract only the main category
main_category = df_clean['category'].str.split('|').str[0]

# Count top 10 main categories
top_categories = main_category.value_counts().head(10)

# Create the figure
plt.figure(figsize=(12,6))

# Plot bar chart
top_categories.plot(
    kind='bar',
    edgecolor='black'
)

# Add title and labels
plt.title("Top 10 Product Categories")
plt.xlabel("Category")
plt.ylabel("Number of Products")

# Rotate x-axis labels for readability
plt.xticks(rotation=45, ha='right')

plt.grid(axis='y', alpha=0.3)

plt.tight_layout()

plt.show()
````

##### 3.1 Why did you pick the specific chart?

A bar chart is the best choice for comparing the frequency of categorical variables. It clearly shows which product categories contain the highest number of products.

##### 3.2 What is/are the insight(s) found from the chart?

* A few product categories dominate the dataset.
* The top categories contain significantly more products than the remaining categories.
* The dataset is not uniformly distributed across all categories.

##### 3.3 Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes.

* Businesses can identify high-demand product segments.
* Marketing campaigns can focus on the categories with the largest product presence.
* Categories with fewer products may represent opportunities for expansion or niche marketing strategies.

#### Chart - 4: Distribution of Discounted Price

````python
plt.figure(figsize=(10,5))

plt.hist(
    df_clean['discounted_price'],
    bins=30,
    edgecolor='black'
)

plt.title("Distribution of Discounted Prices")
plt.xlabel("Discounted Price (₹)")
plt.ylabel("Number of Products")

plt.grid(alpha=0.3)

plt.show()
````

##### 4.1 Why did you pick the specific chart?

A histogram is ideal for understanding the distribution of numerical data. It helps identify common price ranges, skewness, and the presence of expensive or budget-friendly products.

##### 4.2 What is/are the insight(s) found from the chart?

* The distribution of discounted prices is right-skewed, with a large concentration of products in the lower price ranges.
* As the discounted price increases, the number of products decreases significantly.
* Only a small proportion of products are priced at the premium end.

##### 4.3 Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes.
* Businesses can understand the dominant pricing segment in the marketplace.
* Sellers can strategically price their products to remain competitive.
* Premium-priced products can be analyzed separately to understand their market performance.

#### Chart - 5: Top 10 Highest Rated Products

````python
# Select the top 10 most reviewed products
top_reviewed = (
    df_clean[['product_name', 'rating_count']]
    .dropna()
    .sort_values(by='rating_count', ascending=False)
    .head(10)
)

# Shorten long product names for better readability
top_reviewed['short_name'] = (
    top_reviewed['product_name']
    .str.slice(0, 45) + "..."
)

# Create the figure
plt.figure(figsize=(12, 7))

# Plot horizontal bar chart
plt.barh(
    top_reviewed['short_name'],
    top_reviewed['rating_count'],
    edgecolor='black'
)

# Add chart title and labels
plt.title("Top 10 Most Reviewed Products")
plt.xlabel("Number of Ratings")
plt.ylabel("Product")

# Display the highest value at the top
plt.gca().invert_yaxis()

plt.grid(axis='x', alpha=0.3)

plt.tight_layout()

plt.show()
````

##### 1. Why did you pick the specific chart?

A horizontal bar chart is ideal for comparing products with the highest number of customer ratings. It clearly ranks products by popularity while accommodating long product names.

##### 2. What is/are the insight(s) found from the chart?

* A small number of products have received significantly more customer ratings than others.
* These products are likely among the most popular and widely purchased items on Amazon.
* The gap between the top-reviewed products and the rest suggests that customer attention is concentrated on a limited set of products.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes.

* Businesses can identify products that have achieved high customer engagement.
* Popular products can be prioritized in marketing campaigns and recommendation systems.
* Analyzing the features of these products can help sellers improve the visibility and sales of similar products.
